```python
import pandas as pd
import statsmodels.api as sm
```

```python
mmm = pd.read_csv('../assets/MMM.csv', parse_dates=True, index_col='Date',)
sp_500 = pd.read_csv('../assets/^GSPC.csv', parse_dates=True, index_col='Date')
```

```python
mmm.head()
```

```python
sp_500.head()
```

```python
# joining the closing prices of the two datasets 
daily_prices = pd.concat([mmm['Close'], sp_500['Close']], axis=1)
daily_prices.columns = ['MMM', '^GSPC']
```

```python
daily_prices.head()
```

```python
# calculate daily returns
daily_returns = daily_prices.pct_change(1)
clean_daily_returns = daily_returns.dropna(axis=0)  # drop first missing row
clean_daily_returns.head()
```

```python
clean_daily_returns['month'] = clean_daily_returns.index.astype(str).str[:6]

monthly_total = clean_daily_returns.groupby('month').sum()
```

```python
monthly_total.head()
```

```python
monthly_returns = clean_daily_returns.resample('M').agg(lambda x: (x + 1).prod() - 1)
```

```python
monthly_returns.head()
```

```python
yearly_returns = clean_daily_returns.resample('Y').agg(lambda x: (x + 1).prod() - 1)
```

```python
yearly_returns.head(10)
```

```python
mmm_monthly_prices = pd.read_csv('../assets/MMM_Monthly.csv', parse_dates=True, index_col='Date',)
mmm_monthly_prices.head()
```

```python
# calculate monthly returns
mmm_monthly_returns = mmm_monthly_prices['Adj Close'].pct_change(1)
clean_mmm_monthly_returns = mmm_monthly_returns.dropna(axis=0)  # drop first missing row
clean_mmm_monthly_returns.head()
```

```python
mmm_yearly_returns = clean_mmm_monthly_returns.resample('Y').agg(lambda x: (x + 1).prod() - 1)
```
